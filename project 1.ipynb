```python
#a basic program to fetch website content using requests and BeautifulSoup


import requests
from bs4 import BeautifulSoup
from openai import OpenAI

MODEL="llama3.2"
client = OpenAI(base_url="http://localhost:11434/v1", api_key="not_needed") 
def fetch_website_content(url):
    try:
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.text, "html.parser")
        # Remove script & style tags
        for tag in soup(["script", "style"]):
            tag.decompose()
        text = soup.get_text(separator="\n")
        cleaned = "\n".join(line.strip() for line in text.splitlines() if line.strip())
        return cleaned
    except Exception as e:
        return f"ERROR FETCHING WEBSITE: {e}"

system_prompt = "You are a helpful assistant that summarizes website content."
user_prompt = "Summarize the following website content:\n\n{}"  

def summarize_website(url):
    content = fetch_website_content(url)
    full_prompt = user_prompt.format(content)
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": full_prompt}
        ]
    )
    
    return response.choices[0].message.content


#to arrange it in a better format using markdown

def summarize_website_markdown(url):
    content = fetch_website_content(url)
    full_prompt = user_prompt.format(content) + "\n\nPlease format the summary using markdown."
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": full_prompt}
        ]
    )
    
    return response.choices[0].message.content

summarize_website_markdown("https://www.microsoft.com/en-in")
```
