```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from scraper import fetch_website_text
#importing markdown module to print formatted text
from IPython.display import display, Markdown, update_display
load_dotenv(override=True)
MODEL="llama3.2"
client = OpenAI(base_url="http://localhost:11434/v1", api_key="not_needed")

url=input("Enter the website URL: ")
text_contents = fetch_website_text(url)


user_prompt = f"""
Here are the text content of a website I need to understand. Simplify everything and also provide a summary in bullet points at last.
also provide the information as markdown format.
example:
# Heading 1
## Heading 2
- bullet point 1
- bullet point 2
Important Information:
give brief explanation of every detail.
Tips to memorize the information:
"""
user_prompt +="n".join(text_contents.splitlines())
system_prompt="""You are a tutor that helps people find information.
Explain things in simple terms.
Provide a summary in bullet points at last. additionally, highlight any important information.
also provide tips to memorize the information.
"""
response = client.chat.completions.create(
    model=MODEL,
    messages=[{"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt }])
display(Markdown(response.choices[0].message.content))
```
